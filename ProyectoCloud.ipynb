```python
import numpy as np
from typing import Iterable, Optional

def structural_connectivity(data_fname: str, label_fname: str, ref_fname: str,
                            bval_fname: str, bvec_fname: str, mask_fname: str,
                            random_state: Optional[int]=None) -> np.ndarray:
    import gc

    # Paso 1. Abrimos los ficheros y extraemos los datos que necesitamos

    from dipy.io.image import load_nifti_data, load_nifti

    data, _affine, voxel_dim = load_nifti(data_fname, return_voxsize=True)

    ref_volume, ref_affine = load_nifti(ref_fname)
    brain_mask, mask_affine = load_nifti(mask_fname)

    # Paso 2. Registro, en el atlas dado, de la máscara cerebral y la parcelación

    from dipy.align.imwarp import SymmetricDiffeomorphicRegistration
    from dipy.align.metrics import SSDMetric

    sdr = SymmetricDiffeomorphicRegistration(metric=SSDMetric(3), level_iters=(10, 10, 5))

    mapping = sdr.optimize(static=ref_volume, static_grid2world=ref_affine,
                           moving=brain_mask, moving_grid2world=mask_affine)

    labels = load_nifti_data(label_fname)
    label_volume = mapping.transform(labels)

    from scipy.ndimage import zoom

    # Ajustamos el tamaño de la máscara cerebral al de la imagen de parcelación

    labels_shape, wm_shape = np.array(labels.shape), np.array(brain_mask.shape)
    zoom_factor = wm_shape / labels_shape

    label_volume = zoom(label_volume, zoom_factor, order=0)
    volume_shape = np.array(label_volume.shape)

    assert np.all(volume_shape == wm_shape) # Comprobamos que los volúmenes han quedado iguales
    label_volume[brain_mask == 0] = 0 # No consideramos las etiquetas que queden fuera del tejido cerebral

    # Máscara de la materia gris del cerebro
    gray_matter = (label_volume != 0)

    del(labels, brain_mask)
    gc.collect()

    # Paso 3. Leemos los b-valores y b-vectores y hallamos la tabla de gradientes de difusión

    from dipy.io.gradients import read_bvals_bvecs
    from dipy.core.gradients import gradient_table

    bvals, bvecs = read_bvals_bvecs(bval_fname, bvec_fname)
    gtable = gradient_table(bvals, bvecs)

    del(bvals, bvecs)
    gc.collect()

    # Paso 4. Ajustamos un modelo de reconstrucción de tractos nerviosos, para generar después
    #         la tractografía

    from dipy.reconst.csdeconv import ConstrainedSphericalDeconvModel, response_from_mask_ssst

    response, _ratio = response_from_mask_ssst(gtable, data, gray_matter)

    csd_model = ConstrainedSphericalDeconvModel(gtable, response)
    del(gtable, response)
    gc.collect()

    csd_fit = csd_model.fit(data, gray_matter)
    del(data, csd_model)
    gc.collect()

    # Paso 5. Instanciamos un direccionador de tractos nerviosos con los parámetros
    #         obtenidos del modelo

    from dipy.data import default_sphere
    from dipy.direction import DeterministicMaximumDirectionGetter

    csd_dg = DeterministicMaximumDirectionGetter.from_shcoeff(csd_fit.shm_coeff,
                  max_angle=30, sphere=default_sphere)

    # Paso 6. Instanciamos un direccionador de tractos nerviosos con los parámetros
    #         obtenidos del modelo

    from dipy.tracking import utils
    from dipy.tracking.stopping_criterion import BinaryStoppingCriterion

    affine = np.eye(4) # Matriz identidad

    # Delimitamos las ROIs mediante la máscara de materia blanca
    stopping_criterion = BinaryStoppingCriterion(gray_matter)

    from scipy.stats import gmean

    # Calculamos las semillas de la tractografía, los puntos en el espacio de los
    # que van a emanar los tractos nerviosos, en el proceso de calcular la tractografía

    seeds = utils.seeds_from_mask(gray_matter, affine, density=1)
    step_size = gmean(voxel_dim) / 10

    # Paso 7. Generamos la tractografía y hallamos la matriz de conectividad, contando
    #         los tractos nerviosos que hay entre cada par de ROIs

    from dipy.tracking.local_tracking import LocalTracking
    from dipy.tracking.streamline import Streamlines

    # Generador de la tractografía
    streamline_gen = LocalTracking(csd_dg, stopping_criterion, seeds, affine,
                                   step_size, maxlen=200, return_all=False,
                                   random_seed=random_state)

    streamlines = Streamlines(streamline_gen)
    sc = utils.connectivity_matrix(streamlines, affine, label_volume.astype(np.uint16))

    sc = sc[1:, 1:] # La región cero es la materia blanca en el cerebro, por lo que la descartamos
    np.fill_diagonal(sc, 0) # No nos interesa el número de tractos a una misma ROI

    return sc
```

```python
def label_selection(label_fname, selection, result_fname) -> None:

    import nibabel as nib

    labels = nib.load(label_fname)
    data = labels.get_fdata()

    data[np.logical_not(np.isin(data, selection))] = 0
    data = np.searchsorted(selection, data)

    image = nib.Nifti1Image(data, affine=labels.affine, dtype=np.int32)
    nib.save(image, result_fname)
```

```python
import numpy as np

def get_timeseries(atlas_fname: str, func_fname: str, confounds_fname: str=None,
                   mask_fname: str=None) -> np.ndarray:

    from nilearn.maskers import NiftiLabelsMasker

    masker = NiftiLabelsMasker(labels_img=atlas_fname)
    timeseries = masker.fit_transform(func_fname, confounds_fname, mask_fname)

    return timeseries.T
```

```python
import numpy as np
from numba import njit

@njit(parallel=True)
def pearson_correlation(data1: np.ndarray, data2: np.ndarray):

    data1, data2 = np.ravel(data1), np.ravel(data2)

    n = data1.shape[0]
    covariance = np.dot(data1, data2) / n - np.mean(data1) * np.mean(data2)

    return covariance / (np.std(data1) * np.std(data2))

from typing import Optional

@njit(parallel=True)
def functional_connectivity(timeseries: np.ndarray, offset: int, structural_matrix: Optional[np.ndarray]=None):

    lagged_fc = lambda i, j: pearson_correlation(timeseries[i, :-offset], timeseries[j, offset:])

    n_roi = timeseries.shape[0]

    fc = np.zeros((n_roi, n_roi))
    fc_tau = np.zeros((n_roi, n_roi))

    nodes = np.triu_indices(n_roi) if structural_matrix is None else np.nonzero(np.tril(structural_matrix))
    edges = np.stack(nodes, axis=1)

    for i, j in edges:

        fc[i, j] = fc[j, i] = pearson_correlation(timeseries[i, :], timeseries[j, :])
        fc_tau[i, j], fc_tau[j, i] = lagged_fc(i, j), lagged_fc(j, i)

    return fc, fc_tau
```

```python
@njit(parallel=True)
def minmax_norm(data, low=0, high=1):

    min_value = np.min(data)
    minmax_value = np.max(data) - min_value
    assert minmax_value != 0

    return (high - low) * (data - min_value) / minmax_value + low
```

```python
import numpy as np

@njit(parallel=True)
def stuart_landau_oscillator(x, y, a, w, g, gaussian_noise, connectivity_matrix, connectivity_sum):

    local_dynamics = np.multiply(a - np.square(x) - np.square(y), x) - np.multiply(w, y)
    coupling = g * (np.dot(connectivity_matrix, x) - np.multiply(x, connectivity_sum))

    return local_dynamics + coupling + gaussian_noise

@njit(parallel=True)
def hopf_model(t, z, n, a, w, g, noise, m, s):

    eta = noise[:, int(t)]
    x, y = z[:n], z[n:]

    ode_x = stuart_landau_oscillator(x, y, a, w, g, eta, m, s)
    ode_y = stuart_landau_oscillator(y, x, a, -w, g, eta, m, s)

    return np.hstack((ode_x, ode_y))
```

```python
import numpy as np

def ec_gradient_rule(connectivity_matrix, n_roi, rt, offset, n_frames,
                     fc_emp, fc_tauemp, structural_matrix, a, w, g, noise):

    connectivity_sum = np.sum(connectivity_matrix, axis=1)

    from scipy import integrate

    z0 = np.zeros(2 * n_roi)

    ode_sol = integrate.solve_ivp(fun=hopf_model, t_span=(0, n_frames - 1), y0=z0,
                  method='DOP853', t_eval=range(n_frames),
                  args=(n_roi, a, w, g, noise, connectivity_matrix, connectivity_sum))

    assert ode_sol.status == 0

    x_sol, y_sol = ode_sol.y[:n_roi], ode_sol.y[n_roi:]
    fc, fc_tau = functional_connectivity(x_sol, offset, structural_matrix)

    return fc_emp - fc + fc_tauemp - fc_tau, x_sol, fc, fc_tau
```

```python
def power_spectrum_filter(tsample, f_pass, max_attenuation, decay=1.5):

    f_sample, f_nyquist = 1 / tsample, 1 / (2 * tsample)
    f_stop = (f_pass[0] / decay, f_pass[1] * decay)

    nb_pass, nb_stop = np.array(f_pass) / f_nyquist, np.array(f_stop) / f_nyquist

    from scipy import signal

    N, Wn = signal.buttord(nb_pass, nb_stop, gpass=max_attenuation, gstop=75,
                           fs=f_sample)

    return signal.butter(N, Wn, btype='bandpass', fs=f_sample)

def power_spectral_density(timeseries, tsample, f_pass,
                           max_attenuation=None, filter_params=None):

    f_sample = 1 / tsample
    n_sample = timeseries.shape[1]
    no_params = filter_params is None

    if no_params:
        filter_params = power_spectrum_filter(tsample, f_pass,
                                              max_attenuation=max_attenuation)
    from scipy import signal

    nb_timeseries = signal.lfilter(*filter_params, timeseries, axis=1)
    filtered = signal.welch(nb_timeseries, nperseg=n_sample, fs=f_sample, axis=1)

    return (*filtered, filter_params) if no_params else filtered

def hopf_local_bifurcation(ft, pt, t_sample, f_narrowband):

    f_nyquist = 1 / (2 * t_sample)

    from scipy import integrate

    mask_num = np.logical_and(f_narrowband[0] <= ft, ft <= f_narrowband[1])
    mask_den = np.logical_and(f_narrowband[0] <= ft, ft <= f_nyquist)

    b_num = integrate.simps(pt[:, mask_num], ft[mask_num])
    b_den = integrate.simps(pt[:, mask_den], ft[mask_den])

    return b_num / b_den
```

```python
import numpy as np

@njit(parallel=True)
def l21_norm(x1, x2, y1, y2):

    l2_norm = np.sqrt(np.square(x1 - x2) + np.square(y1 - y2))
    return np.mean(l2_norm)

@njit(parallel=True)
def euclidean_norm(v):
    return np.sqrt(np.sum(np.square(v)))

@njit(parallel=True)
def weight_norm(v, g, delta=None):

    v_mag = euclidean_norm(v)

    v_unit = v if v_mag == 0 else v / v_mag
    v_norm = g * v_unit

    if delta is None:
        v_delta = None
    else:
        g_delta = np.dot(delta, v_unit)
        v_delta = delta - np.dot(g_delta, v_unit)

    return v_norm, v_delta

def cosine_annealing(lr_min, lr_max, t_cur, t):

    from math import cos
    return lr_min + 0.5 * (lr_max - lr_min) * (1 + cos(t_cur / t))

from typing import Callable, Optional

def ec_gradient_descent(lr_hopf: float, lr_bifurcation: float, n_iter: int,
                        t_emp, rt, offset, fc_emp, fc_tauemp,
                        f_emp, p_emp, f_narrowband, p_filter_params,
                        structural_connectivity: Optional[np.ndarray],
                        max_connectivity, w, g, beta):

    from IPython.display import clear_output

    n_roi, n_frames = t_emp.shape

    zero = np.zeros((n_roi, n_roi))
    zero_row = np.zeros(n_roi)

    b_emp = hopf_local_bifurcation(f_emp, p_emp, rt, f_narrowband)
    a = zero_row

    noise = beta * np.random.normal(0, beta, size=t_emp.shape)

    ec = zero if structural_connectivity is None else structural_connectivity
    ec_gradient = zero

    from collections import namedtuple
    Distance = namedtuple('Dist', ['iter', 'dist', 'ec'])

    dist_max = 3 # Distancia máxima
    min_dist = Distance(n_iter - 1, dist_max, ec)

    for i in range(n_iter):

        clear_output()
        print('Iteración %d/%d (%.1f%% completado)' % (i + 1, n_iter, i * 100 / n_iter))

        lr_h = cosine_annealing(lr_min=lr_hopf[0], lr_max=lr_hopf[1], t_cur=i, t=n_iter)

        ec_gradient, timeseries, fc, fc_tau = ec_gradient_rule(ec, n_roi, rt, offset,
                                                  n_frames, fc_emp, fc_tauemp,
                                                  structural_connectivity,
                                                  a, w, g, noise)

        ec, ec_gradient = weight_norm(ec, max_connectivity, ec_gradient)
        ec += lr_h * ec_gradient
        ec *= max_connectivity / np.max(ec)

        dist = l21_norm(fc, fc_emp, fc_tau, fc_tauemp)

        if dist < min_dist.dist:
            min_dist = Distance(i, dist, ec.copy())

        lr_b = cosine_annealing(lr_min=lr_bifurcation[0], lr_max=lr_bifurcation[1],
                                t_cur=i, t=n_iter)

        f, p = power_spectral_density(timeseries, rt, f_narrowband, filter_params=p_filter_params)
        b = hopf_local_bifurcation(f, p, rt, f_narrowband)
        a += lr_b * (b_emp - b)

    clear_output()
    print('Iteración %d/%d (100%% completado)' % (n_iter, n_iter))

    if min_dist.iter == n_iter - 1:
        import warnings
        from sklearn.exceptions import ConvergenceWarning

        warnings.warn('No se ha alcanzado la convergencia', ConvergenceWarning)

    return min_dist.ec
```

```python
def angular_velocity(ft, pt):

    peak_mask = np.argmax(pt, axis=1)
    f_intrinsic = ft[peak_mask]

    from math import pi
    return 2 * pi * f_intrinsic
```

```python
def effective_connectivity(fmri_timeseries, fmri_rt, tau=2, max_attenuation=3.0,
                           negative_cutoff=True, structural_matrix=None,
                           max_connectivity=0.2, lr_hopf=(1e-9, 1e-3),
                           lr_bifurcation=(1e-9, 1e-2), n_iter=100,
                           f_narrowband=(0.04, 0.07), g=5.4, beta=0.02):

    n_roi = fmri_timeseries.shape[0]

    assert tau >= fmri_rt
    offset = round(tau / fmri_rt)
    amp_factor = 10 ** (-max_attenuation / 10)

    fmri_timeseries = minmax_norm(fmri_timeseries, low=0, high=amp_factor)
    connectivity_matrix = None

    if structural_matrix is not None:
        connectivity_matrix = minmax_norm(structural_matrix, low=0, high=max_connectivity)

    fc_emp, fc_tauemp = functional_connectivity(fmri_timeseries, offset, structural_matrix)
    ft, pt, p_filter_params = power_spectral_density(fmri_timeseries, fmri_rt, f_narrowband,
                                                     max_attenuation=max_attenuation)
    w = angular_velocity(ft, pt)

    connectivity_matrix = ec_gradient_descent(lr_hopf, lr_bifurcation, n_iter,
                                              fmri_timeseries, fmri_rt, offset,
                                              fc_emp, fc_tauemp, ft, pt,
                                              f_narrowband, p_filter_params,
                                              connectivity_matrix, max_connectivity,
                                              w, g, beta)
    if negative_cutoff:
        connectivity_matrix[connectivity_matrix < 0] = 0

    if structural_matrix is None:
        np.fill_diagonal(connectivity_matrix, 0)
    else:
        connectivity_matrix[structural_matrix == 0] = 0

    return connectivity_matrix
```
